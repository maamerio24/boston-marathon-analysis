```python
import pandas as pd
import numpy as np
import datetime
```

```python
marathon = pd.read_csv('marathon_results_2017 - marathon_results_2017.csv')
df = pd.DataFrame(data=marathon, index=None)
df
```

```python
# Question 1
# Show a representation of the percentage of runners grouped according to their age range.
```

```python
bin_edges = [15, 19, 29, 39, 49, 59, 69, 200]
```

```python
bin_labels = ['15-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']
```

```python
df['Age Division'] = pd.cut(df.Age, bin_edges, labels=bin_labels)
```

```python
percentage_of_runners_by_age = df.groupby('Age Division').Bib.count()/26410
percentage_of_runners_by_age
```

```python
#Question 2
#Find the average finish time by gender

df['Official Time'] = pd.to_datetime(df['Official Time'])
df['Minutes'] = df['Official Time'].dt.hour * 60 + df['Official Time'].dt.minute + df['Official Time'].dt.second/60
```

```python
average_finish = df.groupby('M/F').Minutes.sum()/df.groupby('M/F').Minutes.count()/60
average_finish
```
